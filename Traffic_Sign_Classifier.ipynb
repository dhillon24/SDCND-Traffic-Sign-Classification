### Data preprocessing techniques

The dataset consists of 43 classes with a distribution as shown in the first histogram. Traffic sign images are captured under varied lighting and perspectives, some are even blurred or occluded. This calls for designing a robust system for classification purposes. The images were preprocessed by first converting into the YUV colorspace, then using the Contrast Limited Adaptive Histogram Equalization (CLAHE) technique on the Y channel and then converting back to the RGB colorspace. Y, being the luma component (brightness), was equalized across every picture as the exposure varies starkly across the entire dataset. This difference in exposure affects the classification performance if not accounted for. The CLAHE algorithm was chosen as it does histogram equalization in patches and adaptively compensates for regions where intesity varies widely (as compared to normal histogram equalization which doesn't work well in images with both overexposed and underexposed regions). The dataset was not converted to grayscale as color is a prominent feature of traffic signs and I wanted the network to associate color with different signs.  

```python
### Define your architectur

import tensorflow as tf

EPOCHS = 50
BATCH_SIZE = 256

'''
def batch_normalization_custom(x,layer,epsilon=1e-12):
    if layer == 'conv':
        batch_mean, batch_var = tf.nn.moments(x,[0,1,2])
    elif layer == 'full':    
        batch_mean, batch_var = tf.nn.moments(x,[0])
    else:
        raise('Specify correct layer')
        
    return tf.nn.batch_normalization(x,batch_mean,batch_var,None,None,epsilon) 
'''

def batch_normalization_custom(x,is_training):
    return tf.contrib.layers.batch_norm(x, is_training=is_training, updates_collections = None, center=True, 
                                        scale=True,epsilon=1e-12,decay=0.9,reuse=False, trainable=True)


def dropout_custom(x,keep,is_training):
    if is_training == True:
        return tf.nn.dropout(fc1, keep)
    else:
        return x
```

### Data Set Split and Augmentation

The validation set was constituted from 20% of the original testing set. Brightness normalization as described previously was done for all sets. Additional data was generated from the new training set through perspective distortions. Perspective distortions are suitable for generating additional data as they resemble the different perspectives in which the traffic signs maybe actually be captured by a camera. However, the distortions were somewhat constrained as excessive distortions or rotations may reverse the meaning of certain signs like 'right turn' or 'left turn'. Additional data was generated to improve generalization performance and to ensure that each class has at least a 1000 samples. Classes were not equalized as the original distribution of classes could represent the a-priori probability of occurence of different traffic signs, so some semblence of the original class distribution was maintained. The new dataset is visualized in the above figures.

```python
from tensorflow.contrib.layers import flatten

def DCNN(x,is_training):    
    # Arguments used for tf.truncated_normal, randomly defines variables for the weights and biases for each layer
    mu = 0
    sigma = 0.1
    epsilon=1e-12
    
    # Layer 1: Convolutional. Input = 32x32x3. Output = 30x30x32.
    conv1_W = tf.Variable(tf.truncated_normal(shape=(3,3,3,32),mean=mu,stddev=sigma))
    conv1_b = tf.Variable(tf.zeros(32))
    conv1 = tf.nn.conv2d(x,conv1_W, strides = [1,1,1,1], padding = 'VALID') + conv1_b
    
    # Batch normalization and Activation.
    
    conv1 = batch_normalization_custom(conv1,is_training)
    
    conv1 = tf.nn.relu6(conv1)
    
    # Layer 2: Convolutional. Input = 30x30x32. Output = 28x28x64.
    conv2_W = tf.Variable(tf.truncated_normal(shape=(3,3,32,64),mean=mu,stddev=sigma))
    conv2_b = tf.Variable(tf.zeros(64))
    conv2 = tf.nn.conv2d(conv1,conv2_W, strides = [1,1,1,1], padding = 'VALID') + conv2_b
    
    # Batch normalization and Activation.
    
    conv2 = batch_normalization_custom(conv2,is_training)
    
    conv2 = tf.nn.relu6(conv2)
    
    # Pooling. Input = 28x28x64. Output = 14x14x64.

    conv2 = tf.nn.avg_pool(conv2,ksize=[1,2,2,1],strides=[1,2,2,1],padding='VALID')
    
    conv2 = batch_normalization_custom(conv2,is_training)
        
    # Layer 3: Convolutional. Input = 14x14x64. Output = 12x12x128.
    
    conv3_W = tf.Variable(tf.truncated_normal(shape=(3, 3, 64, 128), mean = mu, stddev = sigma))
    conv3_b = tf.Variable(tf.zeros(128))
    conv3   = tf.nn.conv2d(conv2, conv3_W, strides=[1, 1, 1, 1], padding='VALID') + conv3_b
      
    # Batch normalization and Activation.
    
    conv3 = batch_normalization_custom(conv3,is_training)
    
    conv3 = tf.nn.relu6(conv3)
    
    # Layer 4: Convolutional. Input = 12x12x128. Output = 10x10x256.
    
    conv4_W = tf.Variable(tf.truncated_normal(shape=(3, 3, 128, 256), mean = mu, stddev = sigma))
    conv4_b = tf.Variable(tf.zeros(256))
    conv4   = tf.nn.conv2d(conv3, conv4_W, strides=[1, 1, 1, 1], padding='VALID') + conv4_b
      
    # Batch normalization and Activation.
    
    conv4 = batch_normalization_custom(conv4,is_training)
    
    conv4 = tf.nn.relu6(conv4)
    
    # Layer 5: Convolutional. Input = 10x10x256. Output = 8x8x512.
    
    conv5_W = tf.Variable(tf.truncated_normal(shape=(3, 3, 256, 512), mean = mu, stddev = sigma))
    conv5_b = tf.Variable(tf.zeros(512))
    conv5   = tf.nn.conv2d(conv4, conv5_W, strides=[1, 1, 1, 1], padding='VALID') + conv5_b
      
    # Batch normalization and Activation.
    
    conv5 = batch_normalization_custom(conv5,is_training)
    
    conv5 = tf.nn.relu6(conv5)
    
    # Pooling. Input = 8x8x512. Output = 4x4x512.

    conv5 = tf.nn.avg_pool(conv5, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')
    
    conv5 = batch_normalization_custom(conv5,is_training)
    
    # Flatten. Input = 4x4x512. Output = 8192.
    
    fc0 = flatten(conv5)
    
    # Layer 5: Fully Connected. Input = 8192. Output = 1024.
    
    fc1_W = tf.Variable(tf.truncated_normal(shape=(8192,1024), mean = mu, stddev = sigma))
    fc1_b = tf.Variable(tf.zeros(1024))
    fc1 = tf.matmul(fc0,fc1_W)+fc1_b
    
    # Batch normalization and Activation.
    
    fc1 = batch_normalization_custom(fc1,is_training)
    
    fc1 = tf.nn.relu6(fc1)
    
    fc1 = dropout_custom(fc1, 0.5, is_training)
    
    # Layer 6: Fully Connected. Input = 1024. Output = 256.
    
    fc2_W  = tf.Variable(tf.truncated_normal(shape=(1024, 256), mean = mu, stddev = sigma))
    fc2_b  = tf.Variable(tf.zeros(256))
    fc2    = tf.matmul(fc1, fc2_W) + fc2_b
    
    # Batch normalization and Activation.
    
    fc2 = batch_normalization_custom(fc2,is_training)
    
    fc2 = tf.nn.relu6(fc2)
    
    fc2 = dropout_custom(fc2, 0.8, is_training)
    
    # Layer 7: Fully Connected. Input = 256. Output = 43.
    
    fc3_W  = tf.Variable(tf.truncated_normal(shape=(256, 43), mean = mu, stddev = sigma))
    fc3_b  = tf.Variable(tf.zeros(43))
    logits = tf.matmul(fc2, fc3_W) + fc3_b
    
    L2_loss = tf.nn.l2_loss(fc3_W)
    
    return logits, L2_loss
```

### Neywork Architecture

A Deep Convolutional Neural Network (DCNN) architecture was designed. The details of the architecture are given below:

## Input
The DCNN architecture accepts a 32x32x3 image as input

### Architecture
**Layer 1: Convolutional 1**: Output - 30x30x32, Filter Size - 3x3

**Batch Normalization**

**Activation**: Relu6 

**Layer 2: Convolutional 2**: Output - 28x28x64, Filter Size - 3x3

**Batch Normalization**

**Activation**: Relu6

**Average Pooling**: Output - 14x14x64, Filter Size - 3x3

**Batch Normalization**

**Layer 3: Convolutional 3**: Output - 12x12x128, Filter Size - 3x3

**Batch Normalization**

**Activation**: Relu6 

**Layer 4: Convolutional 4**: Output - 10x10x256, Filter Size - 3x3

**Batch Normalization**

**Activation**: Relu6

**Layer 5: Convolutional 5**: Output - 8x8x512, Filter Size - 3x3

**Batch Normalization**

**Activation**: Relu6

**Average Pooling**: Output - 4x4x512

**Batch Normalization**

**Flatten**: Output - 8192

**Layer 6: Fully Connected 1**:Output - 1024

**Batch Normalization**

**Activation**: Relu6

**Dropout**: Keep Probability = 0.5

**Layer 7: Fully Connected 2**: Output - 256

**Batch Normalization**

**Activation**: Relu6

**Dropout**: Keep Probability = 0.8

**Layer 8: Fully Connected 3**: Output - 43

### Output
The output is 43 logits which are further processed 

```python
### Train the model

x = tf.placeholder(tf.float32, (None, 32, 32, 3))
y = tf.placeholder(tf.int32, (None))
is_training = tf.placeholder(tf.bool, name='MODE')
one_hot_y = tf.one_hot(y, 43)



rate = 0.001
beta = 1e-5

logits, L2_loss = DCNN(x,is_training)
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(logits, one_hot_y)
loss_operation = tf.reduce_mean(cross_entropy) + beta*L2_loss
optimizer = tf.train.AdamOptimizer(learning_rate = rate)
training_operation = optimizer.minimize(loss_operation)


correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
accuracy_operation = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
saver = tf.train.Saver()

def evaluate(X_data, y_data, mode):
    num_examples = len(X_data)
    total_accuracy = 0
    sess = tf.get_default_session()
    for offset in range(0, num_examples, BATCH_SIZE):
        batch_x, batch_y = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
        accuracy = sess.run(accuracy_operation, feed_dict={x: batch_x, y: batch_y, is_training:mode})
        total_accuracy += (accuracy * len(batch_x))
    return total_accuracy / num_examples



with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    num_examples = len(X_train)
    
    print("Training...")
    print()
    for i in range(EPOCHS):
        X_train, y_train = shuffle(X_train, y_train)
        for offset in range(0, num_examples, BATCH_SIZE):
            end = offset + BATCH_SIZE
            batch_x, batch_y = X_train[offset:end], y_train[offset:end]
            sess.run(training_operation, feed_dict={x: batch_x, y: batch_y, is_training:True})
            
        train_accuracy = evaluate(X_train, y_train, mode=True)
        validation_accuracy = evaluate(X_validation, y_validation, mode=False)
        print("EPOCH {} ...".format(i+1))
        print("Training Accuracy = {:.3f}".format(train_accuracy))
        print("Validation Accuracy = {:.3f}".format(validation_accuracy))
        print()
        
        if validation_accuracy > 0.998:
            saver.save(sess, 'dcnn')
            print("Model saved")
 

           
```

### Training Methods

The model was trained using the Adam optimizer, using a batch size of 256, 50 epochs and a learning rate of 0.001. L2 regularization was done for the output layer weights using a beta of 1e-5. Dropout was used before the final and penultimate layer with keep probabilties of 0.8 and 0.5 respectively. An internal batch normalization parameter named decay was set as 0.9 (helps in improving validation set performance when training set performance is good). Average pooling and Relu6 activation function was used. An boolean variable named 'is_training' was created to deactivate dropout and use estimated population mean/variance for batch normalization during inference. 

### Solution Approach

I followed the traditional convolutional neural network architecture guidelines consiting of convolutional layers interspersed by pooling layers and a few fully connected layers at the end but gave it a little spin of my own. The arhitecture followed the approach of reducing the height and width of layers while increasing their depth along the architecture. However, multiplte adjacent convolutional layers were used - two in first set and three in the second set, each having 3x3 filters to give a larger view of the input volume before the destructive pooling layer. Average pooling was used instead of max pooling as it showed better performance. Batch normalization accelerated the learning process such that an accuracy of 99.0% in 50 epochs was achieved in under an hour on a GTX 970M. A crude form of early stopping was used to prevent overfitting. In constrast to popular opinion on its use with batch normalization, dropout when used only before two of the fully connected layers improved the generalization performance. A possible explanation of increased performance with dropout is that it forces the network to not rely on only a few convolutional features for classification. L2 regularization was further used to prevent overfitting and ensure that the magnitude of the final layer weights is more evenly distributed.
